```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join, splitext
```

```python
voci = ['Spesa_1', 'Spesa_2', 'Spesa_3', 'Spesa_4', 'Spesa_5',
        'Spesa_6', 'Spesa_7', 'Spesa_8', 'Spesa_9', 'Spesa_10',
        'Spesa_11', 'Spesa_12', 'Spesa_TOT']
spesa_col = pd.Series(voci)
iregmap = {
    1 : "PIEMONTE", 2 : "VALLE_AOSTA", 3 : "LOMBARDIA", 4 : "TRENTINO", 5 : "VENETO",
    6 : "FRIULI", 7 : "LIGURIA", 8 : "EMILIA_ROMAGNA", 9 : "TOSCANA", 10 : "UMBRIA",
    11 : "MARCHE", 12 : "LAZIO", 13 : "ABRUZZO", 14 : "MOLISE", 15 : "CAMPANIA", 16 : "PUGLIA",
    17 : "BASILICATA", 18 : "CALABRIA", 19 : "SICILIA", 20 : "SARDEGNA"
}
```

```python
years = list(xrange(1997, 2017 + 1))
```

### Dati per tutta l'Italia

Qui produco il file ITALIA.csv con i dati di tutta Italia in [1997, 2017].

```python
# Dati complessivi per l'Italia per ciascun anno.
main = pd.read_csv('ITALIA/Quadro_Italia_CSV.csv', sep=';', decimal=",")
reddito = pd.read_csv('ITALIA/reddito_medio.csv')
reddito = reddito[reddito.Anno >= 1997].reset_index(drop=True)
main['Reddito'] = reddito['Reddito']
main.set_index('Anno')
main.to_csv('ITALIA/ITALIA.csv', index=False)
main
```

### Dati per regione

Qui produco un file per ogni regione, con tutti i dati dal 1997 al 2017.

```python
# Genera un file CSV per ciascuna regione con i dati dal 1997 al 2017.
p = None
for region in iregmap.itervalues():
    x = pd.read_csv('ANNO/SOURCE/' + region + '.csv')
    y = pd.read_csv('REDDITO_PER_REGIONE/' + region + '.csv')
    y = y[y.Anno >= 1997].reset_index(drop=True)
    x = ((x.drop([1])).drop(labels=['VUOTA'], axis=1)).reset_index(drop=True)
    x['SPESA'] = spesa_col
    x = x.T
    x = x.drop(x.index[0])
    x.columns = spesa_col
    x = x.reset_index()
    x['Reddito'] = y['Reddito']
    x.rename(columns={'index' : 'Anno'}, inplace=True)
    #x = (x.set_index('SPESA')).T
    #x = x.rename_axis('Anno', axis=1)
    p = x
    x.to_csv('./REGIONE/' + region + '.csv', header=True, index=False, float_format='%.2f')#, index_label='Anno')
p
```

### Dati per anno

Una volta prodotti i dati per regione, posso ricavare quelli per anno.

```python
voci_full = voci + ['Reddito']
```

```python
df = pd.read_csv('./REGIONE/PIEMONTE.csv')
df
```

```python
p = None
sorted_regions = []
for region in iregmap.itervalues(): sorted_regions.append(region)
sorted_regions = sorted(sorted_regions)
for y in years:
    result = pd.DataFrame()
    result['SPESA'] = pd.Series(voci_full)
    result = result.set_index('SPESA')
    for region in sorted_regions:
        df = pd.read_csv('./REGIONE/' + region + '.csv')
        df = df.set_index('Anno')
        result[region] = df.loc[y]
    p = result
    result.to_csv('./ANNO/' + str(y) + '.csv')
```
